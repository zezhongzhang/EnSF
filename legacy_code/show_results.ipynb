```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
temp_result = np.load(f'L96_results.npz')
state = temp_result['state_save']
est = temp_result['est_save']
obs = temp_result['obs_save']
rmse = temp_result['rmse_all']
saving_dims = temp_result['saving_dims']
```

```python
# RMSEß
plt.figure(figsize=(8,4))
plt.plot(rmse)
plt.yscale('log')
plt.grid(which='both')
plt.title('RMSE')
plt.tight_layout()
plt.show()
```

```python
# state estimation
for i in range(len(saving_dims)):
    plt.figure(figsize=(8,2))

    plt.subplot(1,2,1)
    plt.plot(state[:,i],label='state')
    plt.plot(est[:,i],label='est')
    plt.legend()
    plt.title(f'state/est (dim={saving_dims[i]})')
    
    plt.subplot(1,2,2)
    plt.plot(obs[:,i])
    plt.title(f'obs (dim={saving_dims[i]})')
    plt.tight_layout()
    plt.show()
```
